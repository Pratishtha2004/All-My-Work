```python
def create_date_column(df):
    def create_date(row):
        if row['iyear'] == 0 or row['imonth'] == 0 or row['iday'] == 0:
            return np.nan
        else:
            return f"{int(row['iyear'])}/{int(row['imonth']):02d}/{int(row['iday']):02d}"

    df['date'] = df.apply(create_date, axis=1)
    
    return df
```

```python
data = create_date_column(df=imp_cols)
```
